```python
import pandas as pd

data = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/seoul_tree/seoul_tree.txt', sep = "\t", skiprows=1)


data.head()
```

### "ì„œìš¸ì‹œì—ëŠ” êµ¬ë³„ë¡œ í‰ê·  ì•½ 400 ê·¸ë£¨ ì •ë„ì˜ `ì´íŒë‚˜ë¬´`ê°€ ìˆë‹¤"ë¼ëŠ” ê°€ì„¤ì— ëŒ€í•´ ê°€ì„¤ê²€ì •ì„ ì‹œí–‰í•˜ì„¸ìš”.

- ë‹¨, `ì„œìš¸ì‹œ`, `ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ`, `ì‹œì„¤ê´€ë¦¬ê³µë‹¨` ë¶€ë¶„ì˜ ë°ì´í„°ëŠ” ì œì™¸í•©ë‹ˆë‹¤.
- ê°€ì„¤ê²€ì • ê²°ê³¼ì˜ pvalue ê°’ì„ `pv1`ì— ì €ì¥í•˜ì„¸ìš”. (`.pvalue` ì‚¬ìš©)

```python
from scipy import stats

df_tree = data[['ìì¹˜êµ¬','ì´íŒë‚˜ë¬´']]
df_tree['ì´íŒë‚˜ë¬´'] = df_tree['ì´íŒë‚˜ë¬´'].str.replace(',','').astype(int)

df_tree = df_tree[~data['ìì¹˜êµ¬'].isin(['ì„œìš¸ì‹œ', 'ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ','ì‹œì„¤ê´€ë¦¬ê³µë‹¨'])]
tree_ttest = stats.ttest_1samp(df_tree['ì´íŒë‚˜ë¬´'],400)

pv1 = tree_ttest.pvalue

print('p-value = %.5f'%pv1)
print("=====================")
```

### ì´ë²ˆì—ëŠ” 10ê°œì˜ êµ¬ë¥¼ ìƒ˜í”Œë§í•˜ì—¬, ìœ„ì— ëŒ€í•œ ê°€ì„¤ê²€ì •ì„ ë‹¤ì‹œ í•œë²ˆ ì‹œí–‰í•´ë³´ì„¸ìš”.

- ë‹¨, ì•„ë˜ì˜ ì½”ë“œë¥¼ ì°¸ì¡°í•˜ì—¬ sampleì„ ìœ„í•œ seed ëŠ” 123ë¡œ ì„¤ì •í•˜ì„¸ìš”.
- ê°€ì„¤ê²€ì • ê²°ê³¼ì˜ pvalueê°’ì„ `pv2`ì— ì €ì¥í•˜ì„¸ìš”


```python
import numpy as np
np.random.seed(123)

df_tree_10 = np.random.choice(df_tree['ì´íŒë‚˜ë¬´'], size = 10)  # "~"ì™€ " "ì„ ì ì ˆí•œ ê°’ìœ¼ë¡œ ì±„ìš°ì‹œë©´ ë©ë‹ˆë‹¤

tree_10_ttest = stats.ttest_1samp(df_tree_10,400)

pv2 = tree_10_ttest.pvalue

print('p-value = %.5f'%pv2)
```

## 2. Two-Sample T-test ì—°ìŠµ

### "ì„œìš¸ì‹œì˜ êµ¬ë³„ í‰ê·  ëŠí‹°ë‚˜ë¬´ìˆ˜ì™€ ì™•ë²šë‚˜ë¬´ì˜ ìˆ˜ëŠ” ì°¨ì´ê°€ ì—†ë‹¤"ë¼ëŠ” ê°€ì„¤ì— ëŒ€í•´ ê°€ì„¤ê²€ì •ì„ ì‹œí–‰í•˜ì„¸ìš”.

- ë‹¨, `ì„œìš¸ì‹œ`, `ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ`, `ì‹œì„¤ê´€ë¦¬ê³µë‹¨` ë¶€ë¶„ì˜ ë°ì´í„°ëŠ” ì œì™¸í•©ë‹ˆë‹¤.
- ê°€ì„¤ê²€ì • ê²°ê³¼ì˜ pvalueë¥¼ `pv3`ì— ì €ì¥í•˜ì„¸ìš”

```python
df_tree_2 = data[['ìì¹˜êµ¬','ëŠí‹°ë‚˜ë¬´','ì™•ë²šë‚˜ë¬´']]

df_tree_2['ì™•ë²šë‚˜ë¬´'][26] = '0'

df_tree_2['ëŠí‹°ë‚˜ë¬´'] = df_tree_2['ëŠí‹°ë‚˜ë¬´'].str.replace(',','').astype(int)
df_tree_2['ì™•ë²šë‚˜ë¬´'] = df_tree_2['ì™•ë²šë‚˜ë¬´'].str.replace(',','').astype(int)

df_tree_2 = df_tree_2[~data['ìì¹˜êµ¬'].isin(['ì„œìš¸ì‹œ', 'ê³µì›ë…¹ì§€ì‚¬ì—…ì†Œ','ì‹œì„¤ê´€ë¦¬ê³µë‹¨'])]

tree_ttest_2 = stats.ttest_ind(df_tree_2['ëŠí‹°ë‚˜ë¬´'], df_tree_2['ì™•ë²šë‚˜ë¬´']) 

pv3 = tree_ttest_2.pvalue

print('p-value = %.5f'%pv3)
print("======================")
```


### Two-Sample One-tailed testë¥¼ ì‚¬ìš©í•˜ì—¬ **í‰ê·  ëŠí‹°ë‚˜ë¬´ìˆ˜ê°€ ì™•ë²šë‚˜ë¬´ ìˆ˜ë³´ë‹¤ í¬ë‹¤**ë¼ëŠ” ê°€ì„¤ì— ëŒ€í•œ ê°€ì„¤ê²€ì •ì„ ì‹œí–‰í•˜ì„¸ìš”.
